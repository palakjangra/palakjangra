## **Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, confusion_matrix
```

### Load Dataset

```python
df = pd.read_csv('creditcard.csv')

print("Dataset Loaded Successfully")
print("Dataset Shape:", df.shape)

df.head()
```

### Prepare Data

```python
X = df.drop("Class", axis=1)
y = df["Class"]

print("Feature Shape:", X.shape)
print("Target Shape:", y.shape)
```

### Scale data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Data Splitting = train and test

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Data:", X_train.shape)
print("Testing Data:", X_test.shape)
```

### Evaluation Function

```python
def evaluate_model(name, y_test, y_pred, y_prob):
    print(f"\n{name} Results")
    print("-" * 30)

    accuracy = accuracy_score(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_prob)

    print("Accuracy:", accuracy)
    print("ROC-AUC Score:", auc_score)

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    return accuracy, auc_score
```

### Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
lr_prob = lr.predict_proba(X_test)[:, 1]

lr_acc, lr_auc = evaluate_model(
    "Logistic Regression",
    y_test,
    lr_pred,
    lr_prob
)
```

### Random Forest 

```python
rf = RandomForestClassifier(
    n_estimators=50,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
rf_prob = rf.predict_proba(X_test)[:, 1]

rf_acc, rf_auc = evaluate_model(
    "Random Forest",
    y_test,
    rf_pred,
    rf_prob
)
```

### Results

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest"],
    "Accuracy": [lr_acc, rf_acc],
    "ROC_AUC": [lr_auc, rf_auc]
})

results
```

### Model Comparison Graph

```python
results.set_index("Model")[["Accuracy", "ROC_AUC"]].plot(kind="bar")
plt.title("Model Comparison")
plt.ylabel("Score")
plt.ylim(0,1)
plt.show()
```

### ROC Curve

```python
fpr1, tpr1, _ = roc_curve(y_test, lr_prob)
fpr2, tpr2, _ = roc_curve(y_test, rf_prob)

plt.plot(fpr1, tpr1, label="Logistic Regression")
plt.plot(fpr2, tpr2, label="Random Forest")
plt.plot([0,1], [0,1], "--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

### Best Model Comparison

```python
if rf_auc > lr_auc:
    best_model = rf
    best_name = "Random Forest"
else:
    best_model = lr
    best_name = "Logistic Regression"

print("Best Model Selected:", best_name)
```
